```
import numpy as np
from math import sin, cos, sqrt, atan2, radians
```

```
# 거리 구하는 함수.
R = 6373.0

def mydistance(a1,b1,a2,b2):
    x1 = radians(a1)
    y1 = radians(b1)
    x2 = radians(a2)
    y2 = radians(b2)
    dlon = x2 - x1
    dlat = y2 - y1
    a = sin(dlat / 2)**2 + cos(y1) * cos(y2) * sin(dlon / 2)**2 
    c = 2 * atan2(sqrt(a), sqrt(1 - a))    
    return R * c
```

```
# 값을 넣어 거리 측정.
data_xpos = []
data_ypos = []
data_long = []
data_lat = []

for i in range(len(data)):
    data_xpos.append(127.2655601)
    data_ypos.append(36.5041166)
    data_long.append(data['Longitude'][i])
    data_lat.append(data['Latitude'][i])
```

```
data_distance = []
for i,j,k,l in zip (data_xpos,data_ypos,data_long,data_lat):
  data_distance.append(mydistance(i,j,k,l))
```

```
data_distance
```

```
data_distance = pd.DataFrame(data = data_distance) 
data_distance.head() 
```

```
data2 = pd.concat([data, data_distance], axis=1)
data2.head()
```

```
data2.rename(columns = {0:'distance'}, inplace=True)
data2
```
