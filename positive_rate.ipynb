```python
import numpy as np
import pandas as pd
import re
```

```python
df_tweet=pd.read_csv("sample_twitter_data_2020-07-01_to_2020-07-31.csv")
df_tweet.head()
```

```python
a=list(np.array(df_tweet['text'].tolist()))
```

```python
stop_words=['불용어','개소리','또라이','세끼','민주당','강다니엘','더하우스','영생닷컴','김현아','방탈출','새끼','ABC주스','시험일정']
```

```python
test_a=[]
for i in df_tweet['text']:
    te=0
    for j in range(len(stop_words)):
        if stop_words[j] not in i:
            te=te+1
            if te==len(stop_words):
                test_a.append(i)
#print(test_a)
```

```python
def clean_text(a):
    line=[]
    for i in range(len(a)):
        te=re.sub('\W|[a-zA-Z]',' ',a[i])
        line.append(te)
    line2=[]
    for i in range(len(line)):
        past=' '.join(line[i].split())
        line2.append(past)
    line2
    test_a=line2
    return test_a
```

```python
test_a=clean_text(test_a)
```

```python


def cleancr(dic):
    line=[]
    for i in range(len(dic)):
        line.append(re.sub('\s','',dic[i]))
        #line.append(re.sub("http?://",'',dic[i]))
    return line
                   
def matches(dic, d):
    return d.count(dic)
```

```python

f=open("positive.txt",'r')
positive=f.readlines()
f.close()
f=open("negative.txt",'r')
negative=f.readlines()
f.close()
positive=cleancr(positive)
negative=cleancr(negative)
```

```python
N=len(test_a)
res=np.zeros(shape=(N,2) , dtype=np.int8)
for i in range(N):
    d=test_a[i]
    cnt=0
    for j in range(len(positive)):
        dic=positive[j]
        cnt+=matches(dic,d)
    res[i,0]=cnt
for i in range(N):
    d=test_a[i]
    cnt=0
    for j in range(len(negative)):
        dic=negative[j]
        cnt+=matches(dic,d)
    res[i,1]=cnt
res
```

```python
pos_rate=res[:,0]/(res[:,0]+res[:,1])
print(sum(pos_rate))
print(sum(pos_rate)/len(pos_rate)*100,'%')
```
