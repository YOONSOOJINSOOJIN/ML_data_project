```
import json
geo_data = '/content/세종시.geojson'
```

```
geo_data = json.load(open(geo_data, encoding='utf-8'))
```

```
import pandas as pd

apt_multi = pd.read_excel("/content/final_predict (1) (4).xlsx")

center = [36.508566, 127.248136]

m = folium.Map(location=center, zoom_start = 15)

folium.Choropleth(
    geo_data = geo_data, 
    data = apt_multi,
    columns = ('시군구명', '예측집값'),
    key_on = 'feature.properties.EMD_NM',
    fill_color = 'Purples',
    legend_name='예측 집값',
).add_to(m)
m.save("다중선형회귀분석 예측 집값 결과.html")
m
```

```
for i in apt_multi.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = apt.loc[i, '팝업']
    lat = apt.loc[i, 'Latitude']
    lng = apt.loc[i, 'Longitude']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(m)

m.save('진짜다중회귀예측값.html')
m
```
