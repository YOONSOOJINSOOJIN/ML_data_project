```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_excel("/content/2020new_train.xlsx")
df_test = pd.read_excel("/content/2020new_test.xlsx")
```

```python
import numpy as np
from sklearn import tree
```

```python
df_train.head()
```

```python
target = df_train["보증금"].values
features_one = df_train[['층', '정부청사거리','COFIX',
       'SNS긍정률', '전용면적', '도시형생활주택']].values
```

```python
pd.get_dummies(df_train['도시형생활주택'])
```

```python
pd.get_dummies(df_test['도시형생활주택'])
```

```python
from sklearn.tree import DecisionTreeRegressor
my_tree_one = tree.DecisionTreeRegressor()
my_tree_one = my_tree_one.fit(features_one, target)

max_depth = 5
min_samples_split = 5

my_tree_one = tree.DecisionTreeRegressor(max_depth = max_depth, min_samples_split = min_samples_split, random_state = 3)
my_tree_one = my_tree_one.fit(features_one, target)
```

```python
print(my_tree_one.feature_importances_)
```

```python
print(my_tree_one.score(features_one, target))
```

```python
test_features = df_test[['층', '정부청사거리','COFIX',
       'SNS긍정률', '전용면적', '도시형생활주택']].values
```
